```python
#Q.1 Discuss string slicing and provide example??
#String slicing allows you to extract a portion substring of a string by specifying a range of indices.
string[start:stop:step]
start: The index where the slice starts
stop: The index where the slice ends
step: How much the index increments after each step

# Example:
s = "Hello, World!"

# Extract 'Hello'
print(s[0:5])  # Output: Hello

# Extract 'World'
print(s[7:12])  # Output: World

# Reverse the string
print(s[::-1])  # Output: !dlroW ,olleH

# Extract every second character
print(s[::2])  # Output: Hlo ol!

#Q.2 Explain the key features of lists in Python
 #They allow for storing ordered collections of items and are mutable

#Ordered: Lists maintain the order of insertion.
#Mutable: Items in a list can be changed after creation.
#Heterogeneous: Lists can store different data types
#Dynamic Size: The size of a list can change by adding or removing elements.
my_list = [1, "apple", 3.14, True]

#Q.3  Describe how to access, modify, and delete elements in a list with examples
#You can access list elements using indexing.
my_list = [10, 20, 30, 40]

# Access the first element
print(my_list[0])  # Output: 10

# Access the last element
print(my_list[-1])  # Output: 40

#Modifying:
#Lists are mutable, so you can change elements using indexing.
my_list[1] = 25  # Change 20 to 25
print(my_list)  # Output: [10, 25, 30, 40]

#Deleting:
#You can delete elements using the del statement, remove(), or pop() method.
# Using del
del my_list[1]
print(my_list)  # Output: [10, 30, 40]

# Using remove() (removes first occurrence of the value)
my_list.remove(30)
print(my_list)  # Output: [10, 40]

# Using pop() (removes by index)
my_list.pop(0)
print(my_list)  # Output: [40]

#Q.4 Compare and contrast tuples and lists with examples
#Tuples are immutable, meaning once created, they cannot be modified, whereas lists are mutable.
#Tuples are generally used for fixed collections of items, while lists are used when you need a collection that can change over time.
#Tuples have less memory overhead compared to lists, making them slightly faster for fixed data.
# List
my_list = [1, 2, 3]
my_list[0] = 10  # Lists are mutable
print(my_list)  # Output: [10, 2, 3]

# Tuple
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # Error, tuples are immutable

#Q.5 Describe the key features of sets and provide examples of their use
#Key Features of Sets in Python
#Sets are unordered collections of unique items.

#Unordered: Elements are stored without a specific order.
#Unique: No duplicate elements are allowed.
#Mutable: You can add or remove elements from a set.
#Set Operations: Supports mathematical operations like union, intersection, and difference.
my_set = {1, 2, 3, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4} (duplicates are removed)

# Add an element
my_set.add(5)
print(my_set)  # Output: {1, 2, 3, 4, 5}

# Perform set operations
another_set = {3, 4, 5, 6}
print(my_set.union(another_set))  # Output: {1, 2, 3, 4, 5, 6}

#Q.6 Discuss the use cases of tuples and sets in Python programming
#Use Cases of Tuples and Sets in Python Programming
#Tuples: Use tuples when you have a collection of items that should not change, such as coordinates, or when you need fixed, ordered data.
point = (10, 20)  # Coordinate, will not change
#Sets: Sets are useful for situations where uniqueness is required, like removing duplicates from a list or performing set operations (e.g., union, intersection).
unique_items = set([1, 2, 2, 3, 3, 4])  # {1, 2, 3, 4}

#Q.7 Describe how to add, modify, and delete items in a dictionary with examples??
#Adding/Modifying: You can add or modify elements in a dictionary by assigning a value to a key.
my_dict = {"name": "Alice", "age": 25}
my_dict["age"] = 26  # Modify an existing key
my_dict["city"] = "New York"  # Add a new key

#Deleting: You can remove a key-value pair using del or pop().
del my_dict["city"]
print(my_dict)  # Output: {"name": "Alice", "age": 26}

my_dict.pop("age")
print(my_dict)  # Output: {"name": "Alice"}

#Q.8 Discuss the importance of dictionary keys being immutable and provide examples????
#Importance of Dictionary Keys Being Immutable
#Dictionary keys must be immutable to maintain their integrity as part of a hash-based data structure.
# Mutable objects like lists or sets can change after insertion, which would cause errors in the hashing process.
# Valid dictionary keys (immutable types)
my_dict = {1: "one", "name": "Alice", (1, 2): "tuple"}

# Invalid dictionary key (mutable type)
# my_dict = {[1, 2]: "list"}  # Error, lists cannot be dictionary keys
```
